# Import stuff to read MRI data and our stripper name Brook !

```python
import nibabel as nib
import matplotlib.pyplot as plt
from stripper.inference import Brook
```

```python
mask_path = "../example/sub-A00028185_ses-NFB3_T1w_brainmask.nii.gz"
data_path = "../example/sub-A00028185_ses-NFB3_T1w.nii.gz"
```

```python
data = nib.load(data_path).get_fdata()
label = nib.load(mask_path).get_fdata()
```

# Create a stripper !

```python
stripper = Brook()
```

# Extract a slice from your MRI (for example a coronal slice) that correspond to the first dimension of the tensor

```python
example_slice = data[120,:,:]
```

# Strip the head of that skull, beware, it's going to bleed ...

```python
brain,mask = stripper.strip(example_slice,pre_process=False,post_process=False)
```

# See the result ! 

```python
plt.imshow(example_slice,cmap='gray')
plt.show()

plt.imshow(brain,cmap='gray')
plt.show()
plt.imshow(mask,cmap='gray')
plt.show()

```
